## Number of Distinct Vals for Each Column

```python
unique_counts = trn.nunique()

unique_counts
```

soil type and wilderness area vars are categorial -- i'll want to see if they're mutually exclusive, though.

Also, there are some of these variables that only have one unique val, so we can drop those columsn.

```python
drop_vars = np.where(unique_counts < 2)[0]

drop_vars
```

```python
trn.drop(trn.columns[drop_vars], axis = 1, inplace = True)

trn.shape
```

# Checking if Soil Types and Wilderness Areas are Mutually Exclusive

```python
soil_cols = trn.filter(regex = "^Soil").sum(axis = 1)

soil_cols.value_counts()
#ok, so not mutually exclusive
```

```python
wilderness_cols = trn.filter(regex = "^Wilderness").sum(axis = 1)

wilderness_cols.value_counts()
#this also isn't mutually exclusive, but most are only in a single type
```
